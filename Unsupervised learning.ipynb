```python
import pandas as pd
import numpy as np
from apyori import apriori
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:/Users/Home/MainAcademy/groceries.csv", header=None)
df.head()
```

```python
df = df.drop(0, 1)
df = df.drop(np.where(df[1] == 'Item 1')[0])
df
```

```python
df.describe()
```

```python
list_of_sets = []
for i in range(df.shape[0]):
    list_of_sets.append([str(df.values[i,j]) for j in range(df.shape[1]) if pd.notna(df.values[i,j])])
list_of_sets[:3]
```

### Топ-3 продуктів

```python
list_of_products = []
for i in list_of_sets:
    for y in i:
        list_of_products.append(y)
#list_of_products

from collections import Counter
dict_prod_val = Counter(list_of_products)
top_3 = dict(Counter(dict_prod_val).most_common(3))
print(f"Топ 3 продуктів, які купують найчастіше:")
for key, values in top_3.items():
    print(key, values)
```

### Вибір топ-2 пари продуктів

```python
from itertools import combinations

def collect_pairs(list_of_sets):
    pair_counter = Counter()
    for set_ in list_of_sets:
        unique_pairs = sorted(set(set_)) 
        combos = combinations(unique_pairs, 2)
        pair_counter += Counter(combos)
    return pair_counter

pairs = collect_pairs(list_of_sets)

print(f'Найчастіше зустрічається комбінація продуктів\n{pairs.most_common(2)[0]}\n{pairs.most_common(2)[1]}')
```

```python
result = list(apriori(list_of_sets, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2, max_length=2))
result[0]
```

```python
res_list = []
for x in result:
    rule = [x for x in x[0]]
    support = x[1]
    confidence = x[2][0][2]
    lift = x[2][0][3]
    full_set_info = rule, support, confidence, lift
    res_list.append(full_set_info)
res_list
```

```python
df_1 = pd.DataFrame(res_list)
df_1
```

```python
df_1.rename(columns={df_1.columns[0]: 'rule', df_1.columns[1]: 'support', df_1.columns[2]: 'confidence', df_1.columns[3]: 'lift'}, inplace = True)
df_1
```

### топ-2 пари по параметру support

```python
df_1 = df_1.sort_values(by='support', ascending=False)
df_1.head(2)
```

### топ-2 пари по параметру confidence

```python
df_1 = df_1.sort_values(by='confidence', ascending=False)
df_1.head(2)
```

### топ-2 пари по параметру lift

```python
df_1 = df_1.sort_values(by='lift', ascending=False)
df_1.head(2)
```
