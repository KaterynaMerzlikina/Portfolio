```python
df = pd.read_csv(r"C:/Users/Home/MainAcademy/Ecommerce Purchases.csv")

def act_price(x):
    if x > 100:
        x = x * 1.1
    elif 50 < x < 100:
        x = x * 1.2
    else:
        x = x * 1.3
    return round(x, 2)

df['Actual price'] = df['Purchase Price']
df['Actual price'] = df['Actual price'].apply(act_price)
col = ['Company', 'Purchase Price', 'Actual price']
print(f'14. Приклад результатів розрахунку\n{pd.DataFrame(df, columns=col).sample(10)}')

rank_task = df.groupby(['Language'])['Actual price'].mean().sort_values(ascending=False).rank()
print(f"""\n-------
Кожній мові присвоєний ранг, виходячи з середньої суми покупки 
(чим більше середній чек - тим вищий ранг)\n{rank_task}""")
df['Rank'] = df['Language'].map(rank_task)
col = ['Language', 'Actual price', 'Rank']
print(f"""\n15. Приклад результатів розрахунку\n{pd.DataFrame(df, columns=col).head(10)}
-------""")

df['Language'] = df['Language'].astype('category')
df['L_cat'] = df['Language'].cat.codes
col = ['Language', 'L_cat']
print(f'\n16. Приклад кодування змінної\n№1\n{pd.DataFrame(df, columns=col).head(10)}')

df_1 = df
lang_code = {'Language':     {'el': 9, 'pt': 8, 'en': 7, 'de': 6, 'zh': 5, 'it': 4, 'ru': 3, 'fr': 2, 'es': 1,}}
df_1 = df_1.replace(lang_code)
col = ['Language', 'Rank']
print(f'\n16. Приклад кодування змінної\n№2\n{pd.DataFrame(df_1, columns=col).head(10)}')

l_dummies = pd.get_dummies(df['Language'])
df = pd.concat([df, l_dummies], axis=1)
print(f'\n16. Приклад кодування змінної\n№3\n{pd.DataFrame(df).head(2)}')

df.to_csv(r"C:/Users/Home/MainAcademy/Ecommerce Purchases_new.csv")
print('\n17. Базу даних збережено в файл Ecommerce Purchases_new.csv')
```
